# 特征工程

特征工程可以考虑一些交互。首先是不会变动的固有属性，比如二级分类，一级分类，商店，州。从时间维度，我们可以考虑当天、前4周同样的天（比如当天是周一，考虑前四周的周一）、前7天、前30天。从商品的维度，我们考虑按照商品id、该商店该二级品类、该商店该一级品类、商店来进行汇总，之后可以考虑其他店铺或者从州的层面进行汇总。从特殊时间的角度出发，我们可以根据日历统计当天的节日名字，节日类型。还可以考虑距离下一个节日的日期，节日的名字和类型。因为这是从日历获取的内容，所以不算是信息穿越。当天该州的SNAP是否开放。从价格角度，可以考虑前几天是否有提高价格和降低价格行为。
1. 固有属性
    - dept_id
    - cat_id
    - store_id
    - state_id
2. 当天特征
    - price
    - 前七天价格变动幅度
    - 年份
    - 月份
    - 一年的第几个周
    - wm_yr_week是自增序列
    - 当天节日名字
    - 当天节日类型
    - 距离下一个节日的时间
    - 下一个节日的名字
    - 下一个节日的类型
    - snap信息
3. 前一段时间的特征（7天，30天，前4周同一个周天）。统计平均值，标准差，中位数，上升下降的趋势slope
    -  商品销量
    -  店铺销量
    -  品类销量
    -  其他竞品的销量

4. 跨年特征（去年同期连续7天，去年同期周围7周的同一个周天）
    - 商品销量
    - 店铺销量
    - 品类销量
    - 其他竞品销量

没有考虑到的是，需要考虑一些百分比的特征。是否下雨和极端气温也是影响购买的因素，需要外部的数据。以上模型带有的日期其实都是超参数，可以进行调节，但是模型似乎过度复杂了。另一方面，考虑对每天的特征进行指数平滑，越近的样本点权重越大，从而提取新的特征。指数平滑中也有衰减参数，需要进行调整，可能每个大类的衰减参数大小不一样，要分开调整。
